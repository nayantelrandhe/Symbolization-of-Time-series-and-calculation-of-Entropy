```python
import os
import numpy as np
```

```python
#list of threshold values to binarize data

Th=[0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45,0.50] 

#list of word length for final symbolization

Wl=[2,3,4,5]
```

```python
#Importing files

A=os.listdir('**file path**')  
for subfile in A:
    S=np.loadtxt('**file path**/'+subfile, dtype=float)
    mean=np.mean(S)
    S=abs(S-mean)
    Result=[]
    for z in Th:
        D=S*0
        n=0
        #print(S.shape[0]) 
        
#Binarizing the data based on threshold value
        
        while n!=(S.shape[0]-1):
            if S[n]>z:
                D[n]=1
            else:
                D[n]=0
            n=n+1
            
# Converting the sequence into sequence of words
        
        for a in Wl:  
            n=0
            L=[]
            while n!=(S.shape[0]-(a-1)):
                for b in range(1,a+1):
                    if b==1:
                        x=str(int(D[n+(b-1)]))
                    else:
                        x=x+str(int(D[n+(b-1)]))
                L=L+[x]
                n=n+1
            L=np.array(L)
            #print(L.shape)
            
#Calculating Normalised Corrected Shannon Entropy
            
            for m in range(0,L.shape[0]):
                if m==0:
                    Char=[L[0]]
                    F=[1]
                else:
                    z=0
                    y=1
                    for char in Char:
                        if L[m]==char:
                            F[z]=F[z]+1
                            y=0
                        z=z+1
                    if y!=0:
                        Char=Char+[L[m]]
                        F=F+[1]
                    #print(F)
                    z=z+1
                    
            F=np.array(F)
            C=0
            for z in range(0,F.shape[0]):
                if F[z]==0:
                    C+=1
            F=F/(L.shape[0])
            K=F*(np.log2(F))
            K=np.nan_to_num(K)
            K=-1*np.sum(K)
            K=(K+((((F.shape[0])-C)-1)/(2*(F.shape[0])*(0.693))))/(np.log2(F.shape[0])+((F.shape[0]-1)/(2*(F.shape[0])*(0.693))))
            #print(K)
            Result=Result+[K]
    Result=np.array(Result)
    
# Saves result in a csv file with NCSE for different word lengths calculated for each threshold value one after the other 
    
    np.savetxt(subfile+'.csv',Result, delimiter=',')
```
